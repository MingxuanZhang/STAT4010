### Problem A.2

Consider the simple linear regression model $Y_i = \beta_0 + \beta_1x_i + \varepsilon_i$ for $i = 1,...,n$, $\,\,\,\varepsilon_i \sim N(0,\sigma^2)$. In class, we showed that the least squares estimator in matrix-vector form is $\mathbf{\widehat{\beta}} = (\beta_0, \beta_1)^T = (X^TX)^{-1}X^T\mathbf{Y}$. In this problem, I want you to derive the least squares estimators individually, for simple linear regression. This should be easier, since it doesn't require linear algebra!

Least squares requires that we minimize $$f(\beta_0, \beta_1) = \sum^n_{i=1}\bigg(Y_i - [\beta_0 + \beta_1x_i] \bigg)^2$$ over $\beta_0$ and $\beta_1$. 

#### (a) Find the partial derivative of $f(\beta_0, \beta_1)$ with respect to $\beta_0$, and the partial derivative of $f(\beta_0, \beta_1)$ with respect to $\beta_1$. Recall that the partial derivative with respect to $x$ of a multivariate function $h(x,y)$ is calculated by taking the derivative  of $h$ with respect to $x$ while treating $y$ constant.


$$\frac{\partial f}{\partial \beta_0} = -2\sum\bigg(Y_i - [\beta_0 + \beta_1x_i] \bigg) $$

$$\frac{\partial f}{\partial \beta_1} = -2\sum\bigg(Y_i - [\beta_0 + \beta_1x_i] \bigg)x_i $$

#### (b) Use part (a) to find the minimizers,  $\widehat{\beta}_0$ and $\widehat{\beta}_1$, of $f$. That is, set each partial derivative to zero and solve for $\beta_0$ and $\beta_1$. In particular, show

\begin{align*}
\widehat{\beta}_1 &= \frac{\sum^n_{i=1}(x_i - \bar{x})(Y_i -\bar{Y})}{\sum^n_{i=1}(x_i - \bar{x})^2}  \, \, \text{ and }\\ \\
\widehat{\beta}_0 &= \bar{Y} - \widehat{\beta}_1\bar{x}.
\end{align*}

First, we find $\widehat{\beta}_0$: 
\begin{align}
-2\sum\bigg(Y_i - \widehat\beta_0 - \widehat\beta_1x_i \bigg) \overset{set}{=} 0 \implies \sum\bigg(Y_i - \widehat\beta_0 - \widehat\beta_1x_i \bigg) = 0 \\
\implies  \sum Y_i - \sum\widehat\beta_0 - \widehat\beta_1\sum x_i  = 0 \implies n\bar{Y} - n\widehat\beta_0 - \widehat\beta_1n\bar{x}  = 0 \\
\implies \widehat{\beta}_0 = \bar{Y} - \widehat{\beta}_1\bar{x},
\end{align}

where $\widehat\beta_1$ is found below. Note that we added hats once we set the derivative to zero. Now, let's find $\widehat{\beta}_1$.


\begin{align*}
-2\sum\bigg(Y_i - \widehat\beta_0 - \widehat\beta_1x_i \bigg)x_i \overset{set}{=} 0 \implies \sum\bigg(Y_ix_i - \widehat\beta_0x_i - \widehat\beta_1x^2_i \bigg) = 0 \\
\implies \sum Y_ix_i - \widehat\beta_0\sum x_i - \widehat\beta_1\sum x^2_i = 0 \implies \sum Y_ix_i - \widehat\beta_0n\bar{x} - \widehat\beta_1\sum x^2_i = 0 \\
\implies \sum Y_ix_i - (\bar{Y} - \widehat{\beta}_1\bar{x})n\bar{x} - \widehat\beta_1\sum x^2_i = 0 \\
\implies \sum Y_ix_i - n\bar{Y}\bar{x} + \widehat{\beta}_1n\bar{x}^2- \widehat\beta_1\sum x^2_i = 0 \\
\implies \sum Y_ix_i - n\bar{Y}\bar{x} - \widehat{\beta}_1\big(\sum x^2_i - n\bar{x}^2\big) = 0 \\
\implies  \widehat{\beta}_1\big(\sum x^2_i - n\bar{x}^2\big) = \sum Y_ix_i - n\bar{Y}\bar{x} \\
\implies  \widehat{\beta}_1 = \frac{\sum Y_ix_i - n\bar{Y}\bar{x}}{\sum x^2_i - n\bar{x}^2}
\end{align*}
